```python
# import library stegano
from stegano import lsb

# input file gambar dan pesan yang akan disembunyikan
secret = lsb.hide("spongebob.png", "halo saya imam, ini adalah pesan yang saya sembunyikan saat mencoba program stegagnografi.")

# membuat file gambar baru yang telah berisi pesan tersembunyi
secret.save("spongebob_square.png")
```

```python
from stegano import lsb

# menampilkan pesan tersembunyi
print(lsb.reveal("spongebob_square.png"))
```
